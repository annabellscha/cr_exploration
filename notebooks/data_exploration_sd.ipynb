```python

print(len(df))
years = ["2014","2015","2016"]
for year in years:
    mask2 = df['Date_Incorporation'].str.contains(year)
    df = df[~mask2]
    print(len(df))

print(len(df))
# Display the DataFrame after rows with 'UG' have been removed

# Create a mask where the 'company' column contains 'UG'
mask = df['Company_Name'].str.contains('UG')

# Use the negation of the mask to filter out the rows with 'UG'
df = df[~mask]
print(len(df))
print(df)
```

```python
from fuzzywuzzy import fuzz
import pandas as pd
from fuzzywuzzy import process
# Sample data
#add column with name company to a df
detector = df['Company_Name']
# print(detector)
# print(len(detector))

df_matches = pd.read_csv('name_match.csv')
print(df_matches)
#add detector as column to df_matches
df_matches['detector'] = detector
matches = []
print(df_matches)
# Create a new column 'score' that calculates the similarity between 'name1' and 'name2'
for name in df_matches['Organization Name']:
    match = process.extractOne(name, df_matches['detector'])
    print(name, match[0],match[1])
    matches.append((name, match[0], match[1]))  # match[0] is the matched name from names2, match[1] is the score

# Now matches is a list of tuples, where each tuple is (name from names1, best match from names2, score)
for match in matches:
    print(match)
# print(df)
```

```python
#save matches to csv
print(matches)

#put into df
df_matches = pd.DataFrame(matches, columns=['name1', 'name2', 'score'])
print(df_matches)

df_matches.to_csv('name_matches_scores.csv', index=False)
```
