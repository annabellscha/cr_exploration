```python
from nameparser import HumanName
import gender_guesser.detector as gender

#parse the shareholder_name_mapped and extract the first name
def getFirstName(name):
    name = HumanName(name)
    return name.first

def getLastName(name):
    name = HumanName(name)
    return name.last

# Initialize the gender detector
detector = gender.Detector(case_sensitive=False)

# Define a function that takes a name and returns the gender
def get_gender(name):
    # Split the name and take the first part as the first name
    first_name = name.split()[0]
    # Return the detected gender
    return detector.get_gender(first_name)

def get_title(name):
    name = HumanName(name)
    return name.title
# Apply the function to each row in the 'shareholder_name_mapped' column
# and store the result in a new 'gender' column

df_gender_mapping['firstname'] = df_gender_mapping['shareholder_name'].apply(getFirstName)
df_gender_mapping['lastname'] = df_gender_mapping['shareholder_name'].apply(getLastName)
df_gender_mapping['title'] = df_gender_mapping['shareholder_name'].apply(get_title)
df_gender_mapping['gender'] = df_gender_mapping['shareholder_name'].apply(get_gender)
df_gender_mapping
# name = HumanName("Dr. Juan Q. Xavier de la Vega III")
# print(name.first)  # Outputs 'Juan'

df_gender_mapping
```
